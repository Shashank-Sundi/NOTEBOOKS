```python
#Q2
"""Problem Statement 2:
The college bookstore tells prospective students that the average cost of its
textbooks is Rs. 52 with a standard deviation of Rs. 4.50. A group of smart statistics
students thinks that the average cost is higher. To test the bookstoreâ€™s claim against
their alternative, the students will select a random sample of size 100. Assume that
the mean from their random sample is Rs. 52.80. Perform a hypothesis test at the
5% level of significance and state your decision."""

from scipy.stats import norm
mu_claim=52 
std_claim=4.50
n=100
mu_sample=52.8

#H0 : mu=52
#H1 : mu > 52
#right tailed test

alpha=0.05
z =((mu_sample-mu_claim)*(n**0.5))/std_claim
z_critical=norm.ppf(1-0.05)
print(f"The z statistic={z} > z critical={z_critical} , hence we reject the null hypothesis .")
print(f"Thus the average cost of the textbooks is indeed greater than Rs. 52 ")
```

```python
#Q3
"""Problem Statement 3:
A certain chemical pollutant in the Genesee River has been constant for several
years with mean Î¼ = 34 ppm (parts per million) and standard deviation Ïƒ = 8 ppm. A
group of factory representatives whose companies discharge liquids into the river is
now claiming that they have lowered the average with improved filtration devices. A
group of environmentalists will test to see if this is true at the 1% level of
significance. Assume \ that their sample of size 50 gives a mean of 32.5 ppm.
Perform a hypothesis test at the 1% level of significance and state your decision."""

mu=34
std=8
alpha=0.01
n=50
mu_sample=32.5

#H0 : mu<34
#H1 : mu>=34 
# Right tailed test

z_critical=norm.ppf(1-0.01)
z=(mu_sample-mu)*(50**0.5)/std
print(f"The z statistic={z} < z critical={z_critical} , hence we accept the null hypothesis.")
print("Thus,the factory's claim that the the pollutant quantity has been lowered is TRUE")
```

```python
#Q13
"""Problem Statement 13:
Pooled estimate of the population variance
Does the data provide sufficient evidence to conclude that average percentage
increase in the price of sugar differs when it is sold at two different prices?
Population 1: Price of sugar = Rs. 27.50
n1 = 14
x1 = 0.317%
s1 = 0.12%
Population 2: Price of sugar = Rs. 20.00
n2 = 9
x2 = 0.21%
s2 = 0.11%"""

from scipy.stats import t
n1 = 14
mu1 = 0.317
std1 = 0.12

n2 = 9
mu2 = 0.21
std2 = 0.11

# pooled variance t-test
se=((((n1-1)*(std1**2))+((n2-1)*(std2**2)))/(n1+n2-2))**0.5
n_1=((1/n1)+(1/n2))**0.5

t_statistic=(mu1-mu2)/(se*n_1)
t_critical=t.ppf(1-0.05/2,n1+n2-2)
print(f"The t_score {t_statistic} falls within the critical region [ {-t_critical} , {t_critical}] ,hence we reject the null hypothesis that average percentage increase in the price of sugar is SAME when it is sold at two different prices")
```

```python
#Q15
""" Problem Statement 15:
Comparisons of two population proportions when the hypothesized difference is zero
Carry out a two-tailed test of the equality of banksâ€™ share of the car loan market in
1980 and 1995.
Population 1: 1980
n1 = 1000
x1 = 53
ð‘ 1 = 0.53
Population 2: 1985
n2 = 100
x2 = 43
ð‘ 2= 0.53
"""
#p1,p2--proportion of share of a bank in 1980 and 1995 
#x1,x2--no. of success 

from scipy.stats import norm
n1 = 1000
x1 = 53
ð‘1 = 0.53

n2 = 100
x2 = 43
ð‘2= 0.53

#H0: p1-p2=0
#H1: p1-p2!=0

p=(x1+x2)/(n1+n2) #pooled proportion
se=(p*(1-p)/n1 +p*(1-p)/n2)**0.5
z_statistic=(p1-p2)/se
z_critical=norm.ppf(1-0.05/2)
print(f"The z_score {z_statistic} falls outside the critical region [ {-z_critical} , {z_critical}] ,hence we accept the null hypothesis that the proportion of banksâ€™ share of the car loan market in 1980 and 1995 is same.")
```

```python
#Q17
"""
A die is thrown 132 times with the following results: Number turned up: 1, 2, 3, 4, 5, 6
Frequency: 16, 20, 25, 14, 29, 28
Is the die unbiased? Consider the degrees of freedom as ð‘ âˆ’ 1.
"""
#Expected Frequency=132/6=22(if dice is unbiased)

#H0:The die is unbiased
#H1:The die is not unbiased.

from scipy.stats import chisquare , chi2
obs_f=[16, 20, 25, 14, 29, 28]
exp_f=[22,22,22,22,22,22]
chi_score=chisquare(obs_f, exp_f,6-1)
chi_critical=chi2.ppf(1-0.05/2,5)
print(f"The chi_score {chi_score[0]} falls outside the critical region [-{chi_critical} {chi_critical}] ,hence we accept the null hypothesis that the dice is unbiased")
```

```python
#Chi-Square Test of Independence

"""
H0:Gender is independent of voting
H1:Gender and Voting are dependent 
"""
#We need to create the table for expected observations

observed_voted_men=2792
observed_voted_women=3591
observed_not_voted_men=1486
observed_not_voted_women=2131

total_voted=2792+3591
total_not_voted=1486+2131
total_men=2792+1486
total_women=3591+2131

expected_voted_men=(total_voted*total_men)/10000
expected_voted_women=(total_voted*total_women)/10000
expected_not_voted_men=(total_not_voted*total_men)/10000
expected_not_voted_women=(total_not_voted*total_women)/10000

chisquare1=(((observed_voted_women-expected_voted_women)**2)/expected_voted_women)
chisquare2=(((observed_voted_men-expected_voted_men)**2)/expected_voted_men)
chisquare3=(((observed_not_voted_men-expected_not_voted_men)**2)/expected_not_voted_men)
chisquare4=(((observed_not_voted_women-expected_not_voted_women)**2)/expected_not_voted_women)
chisquare=chisquare1+chisquare2+chisquare3+chisquare4

# OR

from scipy.stats import chi2_contingency
obs=np.array([[2792,3591],[1486,2131]])
a=chi2_contingency(obs,False)
print(f"The chi2 statistic = {a[0]} is within the critical region {chi2.ppf(1-0.05,1)}.Hence we reject the null hypothesis; so, gender and voting are dependent.")
```
